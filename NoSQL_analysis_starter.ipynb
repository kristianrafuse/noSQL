### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

Penang_Flavours = establishments.find_one({'BusinessName': 'Penang Flavours'})

degree_search = 0.01
latitude = Penang_Flavours['geocode']['latitude']
longitude = Penang_Flavours['geocode']['longitude']

query = {'geocode.latitude': {'$gt': (latitude - degree_search), '$lt': (latitude + degree_search)},
         'geocode.longitude': {'$gt': (longitude - degree_search),'$lt': (longitude + degree_search)},
         'RatingValue': 5}

sort = {'scores.Hygiene': 1}

#limit results for printing
results = establishments.find(query).sort(list(sort.items())).limit(5)

#total results for DataFrame
df_data = establishments.find(query).sort(list(sort.items()))

# Print the results
for restaurant in results: 
    pprint(restaurant)
```

```python
# Convert result to Pandas DataFrame
hygiene_results_df = pd.DataFrame(list(df_data))
hygiene_results_df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:

# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': 0}}


# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id': "$LocalAuthorityName", 
                          'count': { '$sum': 1 }}}

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': { 'count': -1, '_id': 1 }}

# Put the pipeline together
pipeline = [match_query, group_query, sort_values]

# Run the pipeline through the aggregate method and save the results to a variable
hygiene_results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result

print(len(hygiene_results))

# Print the first 10 results

pprint(hygiene_results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_results_df = pd.DataFrame(hygiene_results)

# Display the number of rows in the DataFrame
print(len(hygiene_results_df))

# Display the first 10 rows of the DataFrame
hygiene_results_df.head(10)
```
